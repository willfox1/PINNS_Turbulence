```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("DNS_29_Periodic_Hills/alph05-4071-3036.dat", skiprows=20)
df = pd.DataFrame(data, columns = ["x","y","u_mean","v_mean","w_mean","p_mean","dissipation_mean","vorticity_mean",\
                                        "uu","vv","ww","uv","uw","vw","pp"])
```

```python
# Take a sample of the data (full data requires 163GiB of ram!)
sampled_df = df.sample(frac=0.1)

x = sampled_df['x'].values
y = sampled_df['y'].values
u = sampled_df['uu'].values
p = sampled_df['p_mean'].values
pp = sampled_df['pp'].values
diss = sampled_df['dissipation_mean'].values
vort = sampled_df['vorticity_mean'].values
umean = sampled_df['u_mean'].values
vmean = sampled_df['v_mean'].values
wmean = sampled_df['w_mean'].values
```

```python
# Create the contour plot.
contour_plot = plt.tricontourf(x, y, u, cmap='viridis')

# Add color bar and labels.
plt.colorbar(contour_plot, label='Velocity (u)')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Velocity Profile Contour Plot')

# Show the plot or save it to a file.
plt.show()
```
