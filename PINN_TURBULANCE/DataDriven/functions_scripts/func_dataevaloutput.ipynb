```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("u,v,p predict data/predictions.dat", skiprows=20)
df = pd.DataFrame(data, columns = ["x","y","vorticity_mean"])
```

```python
# Take a sample of the data
sampled_df = df.sample(frac=1)

x = sampled_df['x'].values
y = sampled_df['y'].values
vorticity_mean = sampled_df['vorticity_mean'].values
```

```python
# Create the contour plot.
contour_plot = plt.tricontourf(x, y, vorticity_mean, cmap='viridis')

# Set the color bar range between 0 and 35
#contour_plot.set_clim(0, 35)

# Add color bar and labels.
color_bar = plt.colorbar(contour_plot, label='vort_mean')
contour_plot.set_clim(0, 35)
plt.xlabel('x Values')
plt.ylabel('y Values')
plt.title('vort_mean contour plot')

# Show the plot or save it to a file.
plt.show()
```
